```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
vocab_size = 5
N_samples = 20
window_size = 2
word_indeces = torch.randint(vocab_size,(N_samples,))
word_contexts = torch.randint(vocab_size+1,(N_samples,window_size*2))
```

```python
emb_dim = 7
w_emb = torch.randn(vocab_size+1,emb_dim)
w_emb[-1,:] = 0
w_out = torch.randn(vocab_size,emb_dim)
```

```python
class myModel(torch.nn.Module):
    
    def __init__(self):
        super(myModel,self).__init__()
        self.emb = torch.nn.Embedding(vocab_size+1,emb_dim,padding_idx=vocab_size)
        self.out = torch.nn.Linear(emb_dim,vocab_size, bias=False)
    
    def forward(self,x):
        emb = self.emb(x)
        out = self.out(emb)
        return out#.permute(0,2,1)
```

Verdadera loss:

```python
model = myModel()
model.load_state_dict({'emb.weight':w_emb.clone(),
                       'out.weight':w_out.clone()})

#wi_reshape = word_indeces.view(N_samples,1).repeat(1,word_contexts.size(1))
scores = model(word_indeces).view(-1,vocab_size,1).repeat(1,1,word_contexts.size(1))
target = word_contexts
print(scores.size())
print(target.size())
lf = nn.CrossEntropyLoss(ignore_index=vocab_size,reduction='none')
loss1 = lf(scores,target)
print(loss1.sum())
print()

scores = model(word_indeces)
target = word_contexts[:,0]
lf = nn.CrossEntropyLoss(ignore_index=vocab_size,reduction='none')
loss2 = lf(scores,target)
print(loss1[:,0] == loss2)
```

```python
def loss1(scores,target):
    s = scores.view(-1,vocab_size,1).repeat(1,1,target.size(1))
    lf = nn.CrossEntropyLoss(ignore_index=vocab_size,reduction='sum')
    #print(lf(s,target))
    return lf(s,target).sum()

def loss2(scores,target):
    
    new_scores = torch.cat((scores, torch.zeros((scores.size(0),1),device=torch.device('cuda:1'))),1)
#     s = 0
#     for i in range()
#     s += new_scores[i,:][word_contexts[i,:]].sum()
    first_term = - new_scores.view(N_samples,vocab_size+1,1)[:,word_contexts][:,0,:,:].squeeze(2).sum(dim=1)
    second_term = 2 * window_size * torch.logsumexp(scores,dim=1)
    return (first_term + second_term).sum()
        
def loss3(scores,target):
    #s = scores.view(-1,vocab_size,1).repeat(1,1,target.size(1))
    lf = nn.CrossEntropyLoss(ignore_index=vocab_size,reduction='none')
    s = 0
    for i in range(2*window_size):
        s1 = lf(scores,target[:,i]).sum()
        s += s1
        #print(s1)
    return s

model = myModel()
model.load_state_dict({'emb.weight':w_emb.clone(),
                       'out.weight':w_out.clone()})
model.to(device=torch.device('cuda:1'))
scores = model(word_indeces.to(device=torch.device('cuda:1')))
target = word_contexts.to(device=torch.device('cuda:1'))

print(loss1(scores,target))
print()
print(loss3(scores,target))
```

```python
model = myModel()
model.load_state_dict({'emb.weight':w_emb.clone(),
                       'out.weight':w_out.clone()})
wi_reshape = word_indeces.view(N_samples,1).repeat(1,word_contexts.size(1))
optimizer = optim.SGD(model.parameters(), lr=1e-1)
model.train()
optimizer.zero_grad() # Llevo a cero los gradientes de la red
scores1 = model(wi_reshape)
lf1 = nn.CrossEntropyLoss(ignore_index=vocab_size,reduction='none')
print(scores1.size(),word_contexts.size())
#scores1_reshape = scores1.view(-1,vocab_size,1).repeat(1,1,word_contexts.size(1))
loss1 = lf1(scores1,word_contexts).sum(dim=0).mean(dim=0)
loss1.backward()

print(loss1)
print(model.emb.weight.grad)
```

```python
x = torch.zeros(N_samples,vocab_size)
for i, idx in enumerate(word_indeces):
    x[i][idx] = 1

W_EMB = w_emb[:-1,:].clone()
W_EMB.requires_grad = True
W_OUT = w_out.clone()
W_OUT.requires_grad = True

optimizer = optim.SGD((param for param in [W_EMB, W_OUT]), lr=1e-1)
optimizer.zero_grad() # Llevo a cero los gradientes de la red
emb = F.linear(x,W_EMB.T)
scores2 = F.linear(emb,W_OUT)
lf2 = nn.CrossEntropyLoss(ignore_index=vocab_size)
loss2 = .25 * (lf2(scores2,word_contexts[:,0]) + lf2(scores2,word_contexts[:,1]) + lf2(scores2,word_contexts[:,2]) + lf2(scores2,word_contexts[:,3]))
loss2.backward() # Calculo los gradientes
print(loss2)
print(W_EMB.grad)
```

```python
import torch
from NLPUtils import *

import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
with open('../15-Training-word-vectors/text8','r') as file:
    corpus = file.read()
    corpus = [corpus.split(' ')]
```

```python
# Hiperparámetros:
method = 'SkipGram'
window_size = 8
embedding_dim = 200
cutoff_freq = 4
batch_size = 256
use_gpu = 1

trainer_200 = Word2vecTrainer(corpus,cutoff_freq=cutoff_freq,lm=method,window_size=window_size,batch_size=batch_size,embedding_dim=embedding_dim,use_gpu=use_gpu)
```

```python
emb_weight = torch.empty()
parameters = {'emb.weight': torch.nn.init.normal_()}
requires_grad = None

trainer_200.InitParameters(params_dict=parameters,requires_grad_dict=requires_grad)
```

Entrenamiento con SGD:

```python
epochs = 1
learning_rate = 1e-3
sample_loss_every = 100
algorithm = 'SGD'

trainer_200.Train(algorithm=algorithm, epochs=epochs, sample_loss_every=sample_loss_every, lr=learning_rate)
```

Entrenamiento con Adam:

```python
epochs = 1
learning_rate = 1e-2
sample_loss_every = 100
algorithm = 'Adam'

trainer_200.Train(algorithm=algorithm, epochs=epochs, sample_loss_every=sample_loss_every, lr=learning_rate)
```

```python
fig, ax = plt.subplots()
ax.plot(trainer_200.loss_history['iter'],trainer_200.loss_history['loss'])
```
