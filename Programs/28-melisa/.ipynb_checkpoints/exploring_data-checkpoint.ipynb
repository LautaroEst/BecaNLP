```python
import pandas as pd
import numpy as np
import os
import re

%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
# Leo los datos:
rs = np.random.RandomState(127361824)
list_parts = np.array(os.listdir('./parts/'))
df = pd.concat([pd.read_csv('./parts/{}'.format(filename)) for filename in rs.choice(list_parts,100,replace=False)])
df.head(5)
```

```python
# Cuento las muestras por país
print('Muestras por país:')
print(df['country'].value_counts())
print()

# Cuento las muestras por idioma, a partir del país
print('Muestras por idioma:')
df_por = df[df['country'] == 'MLB'].reset_index(drop=True)
df_esp = df[df['country'] != 'MLB'].reset_index(drop=True)
print('Portugués:', len(df_por))
print('Español:', len(df_esp))
```

```python
# Longitud de los comentarios:
lenghts = df['review_content'].str.len().value_counts()
plt.bar(lenghts.index,lenghts.values)

# Datos estadísticos:
lenghts.describe()
```

### Idea para buscar el vocabulario
Primero separo por espacios (que es el delimitador más probable) y me fijo si hay alguna palabra rara y poco frecuente. Si es así, probablemente tenga un error o esté formada por dos palabras reales, delimitadas por otra cosa que no sea un espacio, así que ahí tengo otro delimitador. Esto lo repito hasta que algún criterio me diga que ya terminé.

```python
comments = df_esp['review_content']
```

```python
vocab = comments.str.split(r'[\s,;\.\\?!\\)\\(]+').explode().value_counts()
print('Vocab len: ',len(vocab))
vocab[-30:]
```

```python
vocab = comments.str.findall(r'[]+').explode().value_counts()
print('Vocab len: ',len(vocab))
vocab[-30:]
```

```python
pattern = re.compile(r'(\s+)([a-z],)')
string = 'Esto es una, prueba'
pattern.split(string)
```
