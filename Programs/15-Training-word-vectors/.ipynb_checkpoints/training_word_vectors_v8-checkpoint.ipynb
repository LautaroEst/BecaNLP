```python
import torch
from NLPUtils import *
from datetime import datetime

import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
with open('text8','r') as file:
    corpus = file.read()
    corpus = [corpus.split(' ')]
```

```python
#corpus = [['hola', 'soy', 'lauti'],['hola', 'soy', 'guadi'], ['chau', 'soy', 'luna','me','quiero','ir']]

# Modelo de lenguaje:
method = 'CBOW'
window_size = 8
embedding_dim = 200
cutoff_freq = 4


# Parámetros de iteración:
batch_size = 256
use_gpu = 1

trainer_200 = Word2vecTrainer(corpus,cutoff_freq=cutoff_freq,lm=method,window_size=window_size,batch_size=batch_size,embedding_dim=embedding_dim,use_gpu=use_gpu)
```

```python
pretrained = 'vectors-emb200-win8-lmCBOW_2020-01-24-15-08-26.bin'
requires_grad = True

trainer_200.InitEmbeddings(from_pretrained=pretrained, requires_grad=requires_grad)
```

```python
epochs = 1
learning_rate = 5e-3
sample_loss_every = 100

trainer_200.SGDTrain(epochs=epochs, learning_rate=learning_rate, sample_loss_every=sample_loss_every)
```

```python
fig, ax = plt.subplots()
ax.plot(trainer_200.loss_history['iter'],trainer_200.loss_history['loss'])
```

```python
vector_file = 'vectors-emb{}-win{}-lm{}_{}.bin'.format(embedding_dim,
                                                       window_size,
                                                       method,
                                                       datetime.now().strftime("%Y-%m-%d-%H-%M-%S"))
trainer_200.SaveEmbeddings(vector_file)
```

```python
word_list = ['anarchism', 'anarcho', 'man', 'woman', 'love', 'god', 
             'hate', 'king', 'queen', 'italy', 'france', 'paris', 'rome']

trainer_200.GetCloseVectors(word_list, firsts=20)
```
