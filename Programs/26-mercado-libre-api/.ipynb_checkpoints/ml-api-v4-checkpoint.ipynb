# Inspección del dataset

Los datos están en `data_all2.csv`

```python
df_all = pd.read_csv('data_all2.csv')
df_all
```

Fijamos una cantidad de máxima de comentarios por producto

```python
max_comments = 10
df_all.groupby('product_id').apply(lambda x: x.sample(n=max_comments,replace=True)).reset_index(drop=True).drop_duplicates(subset=['comment'])
```

```python
df = pd.DataFrame({'a':list('aaabbbccc'), 'b': [1,2,3,4,5,5,5,6,7]})
df.groupby('a').apply(lambda x: x.sample(n=4,replace=True)).reset_index(drop=True)
```

Cantidad de comentarios en cada categoría y por idioma

```python
mask = df_all['country'] == 'MLB'
print('Comentarios en portugués:')
rates_port = df_all[mask]['review_rate'].value_counts().sort_index()
print(rates_port)
print('Comentarios en español:')
rates_esp = df_all[np.logical_not(mask)]['review_rate'].value_counts().sort_index()
print(rates_esp)
print('Comentarios totales:')
rates_port + rates_esp
```
