```python
data = ['Brasil.csv', 'Venezuela.csv', 'Uruguay.csv', 'Perú .csv', 'Mexico.csv', 
        'Argentina.csv', 'Colombia.csv', 'Chile.csv', 'Costa Rica.csv', 
        'Dominicana.csv', 'Bolivia.csv', 'Panamá.csv', 'Honduras.csv', 
        'Guatemala.csv', 'Ecuador.csv', 'Paraguay.csv', 'El Salvador.csv', 'Nicaragua.csv']

def get_counts(data,val_threshold=0):
    df_all = pd.DataFrame({},index=[1, 2, 3, 4, 5])
    for file in data:
        df = pd.read_csv('./data/' + file)
        df_all[file.split('.')[0]] = df[abs(df['valorization']) >= val_threshold]['rate'].value_counts()
            
    return df_all

df_all = get_counts(data,val_threshold=1)
df_all
```
