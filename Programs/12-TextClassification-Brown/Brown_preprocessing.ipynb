```python
categories = brown.categories()
train_sentences = []
test_sentences = []
train_labels = []
test_labels = []
for c, category in enumerate(categories):
    sentences = brown.sents(categories=category)
    NUM_SAMPLES = len(sentences)
    random_idx = np.random.permutation(NUM_SAMPLES)
    NUM_TRAIN = int(5/6 * NUM_SAMPLES)
    train_sentences.extend([sentences[i] for i in random_idx[:NUM_TRAIN]])
    test_sentences.extend([sentences[i] for i in random_idx[NUM_TRAIN:]])
    train_labels.extend([c] * NUM_TRAIN)
    test_labels.extend([c] * (NUM_SAMPLES - NUM_TRAIN))
```

```python
train_data = pd.DataFrame({'sentence':['<SEP>'.join(sent) for sent in train_sentences], 'label':train_labels})
train_data
```

```python
test_data = pd.DataFrame({'sentence':['<SEP>'.join(sent) for sent in test_sentences], 'label':test_labels})
test_data
```

```python
train_data.to_csv('train.csv')
test_data.to_csv('test.csv')
```
