# Cargamos el dataset

```python
df = pd.read_csv('train.csv',sep = '|')
df.columns = ['Pregunta', 'Intencion']
name2idx = {name:idx for idx, name in enumerate(sorted(list(df['Intencion'].unique())))}
idx2name = {idx:name for idx, name in enumerate(sorted(list(df['Intencion'].unique())))}
df['Intencion'] = df['Intencion'].apply(lambda name: name2idx[name])
df.head()
```

# Clasificación por Naive Bayes

```python
from NLPUtils.classification import k_fold_validation, BagOfNgramsVectorizer, MultinomialNB
from itertools import product
        
ngram_ranges = [(1,1), (1,2), (1,3), (1,4)]
maxs_features = [500, 1000, 5000, 10000, 25000, 50000, 60000, 65000, 70000, 75000, 80000]
alphas = alphas = [.01, .02, .04, .08, .09, .1, .2, .4, .8, 1., 2., 4., 8.]

results = []
for ngram_range, max_features, alpha in product(ngram_ranges,maxs_features, alphas):
    model = MultinomialNB(alpha=alpha)
    vectorizer = BagOfNgramsVectorizer(ngram_range=ngram_range,max_features=max_features,token_pattern=r'\b\w+\b')

    scores = k_fold_validation(model,
                               df,
                               vectorizer,
                               k_folds=10,
                               random_state=0,
                               metrics=['accuracy','f1_macro','balanced_accuracy'])

    result = {'params': (ngram_range, max_features, alpha),
              'accuracy': np.mean([score['accuracy'] for score in scores]),
              'f1_macro': np.mean([score['f1_macro'] for score in scores]),
              'balanced_accuracy': np.mean([score['balanced_accuracy'] for score in scores])}
    results.append(result)
    print('Results:')
    print('ngram_range: {}, max_features: {}, alpha: {}'.format(ngram_range,max_features, alpha))
    print('accuracy: {:.5f}'.format(result['accuracy']))
    print('balanced accuracy: {:.5f}'.format(result['balanced_accuracy']))
    print('f1 score macro: {:.5f}'.format(result['f1_macro']))
    print()
```

```python
max_acc = max([result['accuracy'] for result in results])
max_f1 = max([result['f1_macro'] for result in results])
max_balanced_acc = max([result['balanced_accuracy'] for result in results])
print('Max accuracy: {:.5f}. Max balanced accuracy: {:.5f}. Max f1 macro: {:.5f}.'.format(max_acc,max_balanced_acc,max_f1))
```
