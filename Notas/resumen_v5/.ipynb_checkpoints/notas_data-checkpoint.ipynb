La tarea de modelo de lenguaje consiste en obtener la probabilidad $P(w_1,\ldots,w_N)$ de una serie de palabras $w_1,\ldots,w_N$ pertenecientes a un vocabulario $V$. En principio, esta probabilidad se puede obtener aproximadamente teniendo acceso a las probabilidades condicionales.
