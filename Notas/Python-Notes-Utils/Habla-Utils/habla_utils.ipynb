## Signal:

```python
sig0 = Signal() # Señal vacía.
sig1 = Signal.from_wav('./Práctica 1/1.wav') # Señal leída desde un wav.
sig2 = Signal.cos(f=[440, 880], A=[2, 1], L=44100, Fs=44100, t0=0) # Señal formada a partir de una suma de cosenos.
sig3 = Signal.glotic(F0=125, Fs=44100, t_max=1.,mode='Delta Train') # Señal de pulso glótico.
```

```python
plt.rcParams['figure.figsize'] = (9.0, 4.0)
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,(3,4))

sig1.plot(ax=ax1) # Gráfico de la señal en el tiempo.
sig1.fft_plot(ax=ax2) # Gráfico de la fft de la señal.
sig1.spectrogram(ax=ax3) # Gráfico del espectrograma.

sig1.to_wav('sig1.wav') # Almacena la señal en un wav. La cuantización es la misma que el type de la señal.
sig1.play() # Esta instrucción siempre va al final de la celda. Sirve para poder escuchar la canción.
```
